# 모듈

```python
#matplotlib 대표적인 과학계산용 그래프를 그리는 대표적인 라이브러리이다.
import matplotlib.pyplot as plt 
#넘파이는 파이썬의 대표적인 배열(array) 라이브러리이다.
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

# 데이터

```python
breams_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0] 
breams_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
# 분류 작업을 위해, 빙어 14마리의 데이터
smelts_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0] 
smelts_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

# 데이터 확인

```python
plt.scatter(breams_length, breams_weight) 
plt.xlabel('length') 
plt.ylabel('weight') 
plt.show()
```

```python
plt.scatter(breams_length, breams_weight) 
plt.scatter(smelts_length, smelts_weight) 
plt.xlabel('length') 
plt.ylabel('weight') 
plt.show()
```

# 데이터 합치기 

```python
length = breams_length + smelts_length 
weight = breams_weight + smelts_weight
```

# 머신러닝하기 위한 2차원 리스트 만들기 

```python
fishs_data = [[l, w] for l, w in zip(length, weight)]
print(fishs_data)
```

# 정답 데이터 만들기

```python
fishs_target = [1] * 35 + [0] * 14
print(fishs_target)
```

# K-NN 알고리즘 클래스인 KNeighborsClassifier를 import

```python
kc = KNeighborsClassifier()
```

#  도미를 찾기 위한 기준을 학습시킨다(1이 나온다면 모든 데이터가 정확히 맞혔는 의미이다  ,0.5가 나온다면 절반의 데이터를 맞혔는 의미이다)

```python
kc.fit(fishs_data, fishs_target) 
print(kc.score(fishs_data, fishs_target))
```

# KNeighborsClassifier 클래스의 K값 변경(기본 설정값이 5이다)

```python
kc = KNeighborsClassifier(n_neighbors=49) 
kc.fit(fishs_data, fishs_target) 
print(kc.score(fishs_data, fishs_target)) 
```

# 이재 실습을 해보자


```python
#파이썬 리스트를 넘파이 배열로 바꾸는 과정이다
input_arr = np.array(fishs_data) 
target_arr = np.array(fishs_target)

test_arr = np.array(fishs_data) 
tests_arr = np.array(fishs_target)

# K-NN 모델 생성 및 n_neighbors 설정
kc = KNeighborsClassifier(n_neighbors=5)

# 0 ~ 48의 인덱스 중 임의로 35개의 인덱스를 선택한 다음, 해당 인덱스에 해당하는 원소를 뽑아내는 방식으로 훈련 셋을 만들자.
np.random.seed(42) 
index = np.arange(49) # [0, 1, 2, ... , 48] 
np.random.shuffle(index) 

trains_input = input_arr[index[:35]] 
trains_target = target_arr[index[:35]]
tests_input = test_arr[index[:35]] 
tests_target = tests_arr[index[:35]]

# 만든 훈련 셋과 테스트 셋으로 K-NN 모델을 훈련시켜보자.
kc.fit(trains_input, trains_target) 
print("score 메서드: ", end ="")
print(kc.score(tests_input, tests_target)) 

print("출력값", end ="")
print(kc.predict(tests_input)) 
print("정답값", end ="")
print(tests_target) 
```

# K-NN 알고리즘의 주의점

이전의 실습에서 훈련된 모델에 길이가 25cm이고, 무게가 150g인 생선을 테스트해보자.

```python
print(kc.predict([[25.0, 150.0]]))
```

훈련 셋을 참고하면 도미라고 판단해야 할 텐데 빙어라고 예측한다.


위의 산점도를 보면 도미라고 판단해야 할 것이다. 왜 이런 문제점이 발생을 하는걸까?

```python
distances, indexes = kc.kneighbors([[25, 150]])

plt.scatter(trains_input[:,0], trains_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(trains_input[indexes, 0], trains_input[indexes, 1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

산점도를 그려보면 빙어들의 거리는 굉장히 멀리있지만 도미는 1개 포함되있고 빙어는 4개의 샘플은 모두 빙어임을 알수가 있다 실제의 거리를 출력해보자 

```python
print(distances)
```

데이터를 확인해보니 빙어의 거리는 약 137이고 도비의 거리는 92 나타난다 거리가 이상 하다는것을 알수있다 그이유는 x축은 범위가 좁고, y축은 범위가 넓기 때문이다 
x축 범위를 y축과 동일하게 만들어보자. x축의 범위 또는 y축의 범위를 조절하려면 xlim, ylim 함수를 사용하면 된다. 

```python
plt.scatter(trains_input[:,0], trains_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(trains_input[indexes, 0], trains_input[indexes, 1], marker='D')
plt.xlim((0, 1000))
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

실제로 x는 거리에 영향을 거의 미치지 않는 것을 볼 수 있다. 
그러므로 런파이를 활용해서 x축고y축을 동일하게 만들어보었다

```python
mean = np.mean(trains_input, axis=0)
std = np.std(trains_input, axis=0)
print(mean, std) 
trains_scaled = (trains_input - mean) / std
new = ([25, 150] - mean) / std

plt.scatter(trains_scaled[:, 0], trains_scaled[:, 1])
plt.scatter(new[0], new[1], marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

x축과y축이 같은걸 볼수있다. 다시 이데이터를 활용하여 학습 시켜보자

```python
kc.fit(trains_scaled, trains_target)
tests_scaled = (tests_input - mean) / std

kc.score(tests_scaled, tests_target)
print(kc.predict([new]))
```

```python
distances, indexes = kc.kneighbors([[-0.4, -0.4]])

trains_scaled = (trains_input - mean) / std
new = ([25, 150] - mean) / std

plt.scatter(trains_scaled[:, 0], trains_scaled[:, 1])
plt.scatter(new[0], new[1], marker='^')
plt.scatter(trains_scaled[indexes, 0], trains_scaled[indexes, 1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

x축과y축을 동일하게 해고 다시 학습 시키니 도비로 나오게되었다
